```python
#importar librerias
import pandas as pd
import numpy as np 
import plotly as px
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3 as sql
import datetime as dt
```

```python
conn= sql.connect("db_empleados.db")
```

```python
#cargar bases de datos
#df_out_time=pd.read_csv("https://github.com/vladimir-martin/analitica3_HR/raw/d44db9d7c1014ad6b7a588f49d6505e4a02835cd/out_time.csv")
#df_in_time=pd.read_csv("https://github.com/vladimir-martin/analitica3_HR/raw/d44db9d7c1014ad6b7a588f49d6505e4a02835cd/in_time.csv")
df_out_time=pd.read_csv("out_time.csv")
df_in_time=pd.read_csv("in_time.csv")
```

```python
df_in_time.head()
```

```python
#info de las bases
df_out_time.info(verbose=True)
df_in_time.info()
```

```python
#cambiar nombre a columna por employee id
df_out_time=df_out_time.rename(columns={"Unnamed: 0":"EmployeeID"})
df_in_time=df_in_time.rename(columns={"Unnamed: 0":"EmployeeID"})
```

```python
#convertir a formato fecha
out_date=pd.DataFrame(df_out_time.iloc[:,1:].apply(pd.to_datetime))
in_date=pd.DataFrame(df_in_time.iloc[:,1:].apply(pd.to_datetime))
```

```python
#Realizar resta para sacar tiempo trabajado
working_time=out_date-in_date
#Convertir tiempo trabajdo a segundos
working_time_sec=working_time.apply(lambda x: x.dt.total_seconds())
```

```python
#Sacar horas promedio por empleado
working_time_hr=((pd.DataFrame(working_time_sec.T.mean(skipna=True)).rename(columns={0:"Mean_Work_Hrs"}))/3600)
working_time_hr.insert(0,"EmployeeID",df_in_time["EmployeeID"])
```

```python
### Llevar tablas a base de datos sql
working_time_hr.to_sql("time",conn,if_exists="replace")
```
